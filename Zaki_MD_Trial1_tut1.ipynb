[11] remove water in the model

```python
%%bash
cd /content/grojobs
grep -v HOH 1AKI.pdb > 1AKI_clean.pdb
```

[12] generate the topology file using force field equation specified after -ff

```python
%%bash
cd /content/grojobs
source /usr/local/gromacs/bin/GMXRC
gmx pdb2gmx -f 1AKI_clean.pdb -o 1AKI_processed.gro -water spce -ff amber99sb #you can insert the used force field from the command by using -ff fft name
```

[13] add water box configuration

```python
%%bash
cd /content/grojobs
source /usr/local/gromacs/bin/GMXRC
gmx editconf -f 1AKI_processed.gro -o 1AKI_newbox.gro -c -d 1.0 -bt cubic
```

[14] generate solvation file

```python
%%bash
cd /content/grojobs
source /usr/local/gromacs/bin/GMXRC
gmx solvate -cp 1AKI_newbox.gro -cs spc216.gro -o 1AKI_solv.gro -p topol.top
```
